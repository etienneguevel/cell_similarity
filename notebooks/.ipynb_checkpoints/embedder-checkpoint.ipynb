```python
import os 
import torch
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
import psutil

# Function to get memory usage
def get_memory_usage():
    # Get memory usage in bytes
    memory_usage = psutil.virtual_memory().used
    # Convert bytes to megabytes for easier readability
    memory_usage_mb = memory_usage / (1024 * 1024)
    return memory_usage_mb

# Example usage
memory_usage = get_memory_usage()
print("Memory usage:", memory_usage, "MB")

# Function to get available memory
def get_available_memory():
    # Get available memory in bytes
    available_memory = psutil.virtual_memory().available
    # Convert bytes to megabytes for easier readability
    available_memory_mb = available_memory / (1024 * 1024)
    return available_memory_mb

# Example usage
available_memory = get_available_memory()
print("Available memory:", available_memory, "MB")
```

```python
def find_files(path_dir: str):
    """
    help function to find the files in a directory
    args:
    - path_dir: name of the directory to inspect

    outs:
    - files: dictionnary containing as a key the path of the files found and as a value their name in their directory
    """
    files = {}

    for dirpath, dirnames, filenames in os.walk(path_dir):
        temp = {os.path.join(dirpath, f): f for f in filenames}
        files.update(temp)

    return files
```

## UNI model

UNI is a model that was developped by the Brixham women's hospital that trained a large ViT following DINOv2 method.  
It uses patches from histopathological WSI to make a foundation model (100 M patches from 100k WSI).

```python
import timm
from timm.data import resolve_data_config
from timm.data.transforms_factory import create_transform
from huggingface_hub import login
```

```python
# pretrained=True needed to load UNI weights (and download weights for the first time)
# init_values need to be passed in to successfully load LayerScale parameters (e.g. - block.0.ls1.gamma)
model = timm.create_model("hf-hub:MahmoodLab/uni", pretrained=True, init_values=1e-5, dynamic_img_size=True)
transform = create_transform(**resolve_data_config(model.pretrained_cfg, model=model))
model.eval()
```

### Make the dataloaders

```python
from torchvision import datasets
def prepare(image):
    pixel_values=processor(images=image, return_tensors='pt').pixel_values
    inputs = {}
    inputs["pixel_values"] = pixel_values[0,:,:,:]
    return inputs

image_dataset = datasets.ImageFolder(path_data, prepare)
dataloader = torch.utils.data.DataLoader(image_dataset, batch_size=64, shuffle=True, num_workers=4)
```

```python
%%timeit
with torch.no_grad():
    for x in dataloader:
        inputs, lab = x[0], x[1]
        outputs = model(**inputs).last_hidden_state[:,1:,:]
        break
```

```python
%%timeit
with torch.inference_mode():
    for x in dataloader:
        inputs, lab = x[0], x[1]
        outputs = model(**inputs).last_hidden_state[:,1:,:]
        break
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from tqdm import tqdm

results = torch.tensor([])
labels = torch.tensor([])

model.eval()
with torch.no_grad():
    for i, x in enumerate(tqdm(dataloader)):
        inputs, lab = x[0], x[1]
        outputs = model(**inputs).last_hidden_state[:,1:,:]
        results = torch.concat([results, outputs])
        labels = torch.concat([labels, lab])
        if i == 100:
            break
```

```python
tensor_dict = {"embeddings": results, "labels": labels}
torch.save(tensor_dict, 'embedded_cells.pt')
```
